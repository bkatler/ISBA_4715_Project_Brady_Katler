# Stocks I want to get data for: Boeing (BA), Lockheed Martin (LMT), RTX, and S&P 500 as a reference.

```python
#Importing Relevant Libraries
import json
import requests
import pandas as pd
```

```python
#Documentation from Alpha Vantage API - for Boeing: https://www.alphavantage.co/documentation/
boeing_url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=BA&apikey=RIXL2J3YRWMGH0CS'
boeing_r = requests.get(boeing_url)
boeing_data = boeing_r.json()

print(boeing_data)
```

```python
#Documentation from Alpha Vantage API - for LMT: https://www.alphavantage.co/documentation/
lmt_url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=LMT&apikey=RIXL2J3YRWMGH0CS'
lmt_r = requests.get(lmt_url)
lmt_data = lmt_r.json()

print(lmt_data)
```

```python
#Documentation from Alpha Vantage API - for RTX: https://www.alphavantage.co/documentation/
rtx_url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=RTX&apikey=RIXL2J3YRWMGH0CS'
rtx_r = requests.get(rtx_url)
rtx_data = rtx_r.json()

print(rtx_data)
```

```python
#Documentation from Alpha Vantage API - for SPY: https://www.alphavantage.co/documentation/
spy_url = 'https://www.alphavantage.co/query?function=TIME_SERIES_MONTHLY&symbol=SPY&apikey=RIXL2J3YRWMGH0CS'
spy_r = requests.get(spy_url)
spy_data = spy_r.json()

print(spy_data)
```

```python
#Creating a dlist and appending each date, closing price into the list via dictionaries (Boeing)

boeing_stock_data = []

# Extracting date and closing price for each entry and storing in a list of dictionaries
for date, values in boeing_data['Monthly Time Series'].items():
    # Extracting closing price
    closing_price = values['4. close']
    
    # Creating a dictionary for date and closing price
    entry = {'date': date, 'closing_price': closing_price}
    
    # Appending the dictionary to the list
    boeing_stock_data.append(entry)
```
