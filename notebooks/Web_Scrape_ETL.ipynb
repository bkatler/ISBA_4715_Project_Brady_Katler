```python
#Importing Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#Saving website to URL
url = 'https://space.skyrocket.de/doc_chr/lau2023.htm'
```

```python
#Storing URL data in response
response = requests.get(url)
```

```python
#Storing Response in BeautifulSoup Object
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#Checking to make sure website looks good
soup.prettify()
```

```python
#Storing Main Table as a soup object to parse through
main_table = soup.find('table', attrs={'class':'data'})
```

```python
#Removing extra data
for td in soup.find_all('td'):
    for tag in td.find_all(['a', 'br', 'span']):
        tag.decompose()
```

```python
#Cleaning data
for tr in main_table.find_all('tr'):
    td_texts = [td.text.strip() for td in tr.find_all('td')]
    print(td_texts)
```

```python
#Storing relevant fields in dictionary + preparing to export
launch_data = {
    'launch_id': [],
    'launch_date': [],
    'launch_location': [],
    'launch_notes': []
}

for tr in main_table.find_all('tr'):
    tds = tr.find_all('td')
    if len(tds) >= 6:  # Check if there are at least 6 <td> elements
        launch_data['launch_id'].append(tds[0].text.strip())
        launch_data['launch_date'].append(tds[1].text.strip())
        launch_data['launch_location'].append(tds[4].text.strip())
        launch_data['launch_notes'].append(tds[5].text.strip())
```

```python
#Checking to make sure data was stored correctly
launch_data
```

```python
#Storing satellite data in dataframe
df = pd.DataFrame(launch_data)
```

```python
#Checking Data Frame was stored correctly
df.head()
```

```python
#Exporting to CSV
df.to_csv('launch_data_2024', index=False)
```
